```python
import numpy as np
import pandas as pd
```

```python
def misclassifiedByAllNetworks(row):    
    misclassified = 1
    
    for column in columns:
        if row[column] == row.truth:
            misclassified = 0
            break
            
    return misclassified
```

```python
def correctlyClassifiedByAllNetworks(row):    
    correctly_classified = 1
    
    for column in columns:
        if row[column] != row.truth:
            correctly_classified = 0
            break
            
    return correctly_classified
```

```python
labelName = ["MEL", "NV", "BCC", "AK", "BKL", "DF", "VASC", "SCC"]
```

```python
probabilities = pd.read_csv("DataFrames/probabilities.csv", header = [0,1], index_col = 0)
probabilities = probabilities.dropna()
metadata = pd.read_csv("DataFrames/ISIC_2019_Training_Metadata.csv", index_col = "image")
HAM_metadata = pd.read_csv("DataFrames/HAM10000_metadata.csv", index_col = "image_id")
df = pd.read_csv("DataFrames/label.csv", index_col = "image")
```

```python
probabilities
```

```python
predictions = probabilities.T.groupby("Model").idxmax().T.map(lambda x: x[1] if isinstance(x, tuple) else x)
predictions
```

```python
df_truth = df.idxmax(axis=1)
df_truth.name = "truth"
df_truth = df_truth.to_frame()
df_truth
```

```python
df_prediction = predictions.join(df_truth)
df_prediction
```

```python
columns = df_prediction.columns.drop("truth")
df_prediction["misclassified"] = df_prediction.apply(misclassifiedByAllNetworks, axis=1)
```

```python
df_prediction["correctly_classified"] = df_prediction.apply(correctlyClassifiedByAllNetworks, axis=1)
```

```python
df_prediction_correct = df_prediction.query(" misclassified == 0 ")
df_prediction_all_correct = df_prediction.query(" correctly_classified == 1 ")
```
